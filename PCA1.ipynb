## IMPORT LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
```

```python
rng=np.random.RandomState(1)
X=np.dot(rng.rand(2,2),rng.randn(2,200)).T
X
```

### PLOT OF ORGINAL DATA

```python
data_plot=plt.scatter(X[:,0],X[:,1])
plt.title('PLOT OF DATA')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
Data=pd.DataFrame(X,columns=['X','Y'])
Data
```

```python
sns.set()
sns.lmplot(x='X',y='Y', data=Data,fit_reg=False)
```

```python
print("Shape of orginal data:",Data.shape)
```

### 200 rows and 2 columns

```python
model=PCA(n_components=1)
model
```

```python
model.fit(Data)
Data_1D=model.transform(Data)
Data_1D
```

```python
Data['PCA1']=Data_1D[:,0]
Data
```

```python
print("SHAPE OF REDUCED DATA:",Data_1D.shape)
```
